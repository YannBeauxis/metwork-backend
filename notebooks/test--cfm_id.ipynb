```python
from rdkit import Chem
from rdkit.Chem import rdFMCS, Descriptors, AllChem
```

```python
pz_sm = "[H]\C(=C(\[H])C1=CC=C(O)C=C1)N1C(O)=NC(CCCNC(N)=N)C1=O"
pz_mol = Chem.MolFromSmiles(pz_sm)
pz_mol
```

```python
frag_mol = Chem.MolFromSmiles("CCCC1=NC(=O)N(C#CC2=CC=C([OH2+])C=C2)C1O")
frag_mol
```

```python
mols = (pz_mol, frag_mol)
res = rdFMCS.FindMCS(mols)
res.queryMol
```
